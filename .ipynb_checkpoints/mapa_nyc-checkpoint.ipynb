```python
#trabalhando os dados
for i in range(0,len(result_set["LOCATION"])):
    aux=result_set["LOCATION"][i]
    aux=aux[1:-1]
    aux=aux.split(",")
    aux[0]=float(aux[0])
    aux[1]=float(aux[1])
    folium.Marker(location=[aux[0],aux[1]], popup='acidente de carro '+str(i+1),
    icon=folium.Icon(color='blue', icon='info-sign') ).add_to(folium_map)
#float(aux[0])
folium_map
```
